```python
plt.figure(figsize=(12,4))
sns.countplot(x="job", data= bank, hue ="cons.conf.idx")
plt.title("Education Status and Deposits\n")
plt.show()
```

```python
sns.histplot(x="age", data=bank, kde=True, hue= "y")
plt.title("Age Distribution and Deposits\n")
plt.show()
```
